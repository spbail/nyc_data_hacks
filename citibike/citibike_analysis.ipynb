# Part 1: Data loading and transforming

```python
# Some import stuff
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
%matplotlib inline
plt.rcParams['figure.figsize'] = [14,10]
sns.set()
```

```python
# Read in all the trip data files
# I should put these in a DB to query them more easily later, very tbd
# Limiting these to 2017-2021 for now due to size

concat_df = pd.DataFrame()
min_year = 2017
max_year = 2021

def get_urls():
    for y in range(min_year,max_year+1):
        for m in range(1,13):
            y_m = str(y) + "" + '{0:02d}'.format(m)
            # The filename pattern changed in 2017
            if y < 2017:
                url_pattern = 'https://s3.amazonaws.com/tripdata/{}-citibike-tripdata.zip'
            else:
                url_pattern = 'https://s3.amazonaws.com/tripdata/{}-citibike-tripdata.csv.zip'
            yield url_pattern.format(y_m)

for url in get_urls():
    try:
        print("Downloading {}".format(url))
        file_df = pd.read_csv(url)
        print("    Done, read in {} rows.".format(len(file_df)))
        file_df.head()
        grouped_df = file_df.groupby('bikeid').agg(
            {
                'tripduration': 'sum',
                'starttime': 'min',
                'stoptime': 'max',
                'usertype': 'count'
            }
        ).rename({
            'usertype': 'tripcount',
            'starttime': 'first_ride',
            'stoptime': 'last_ride'
        }, axis=1
        ).reset_index()
        concat_df = concat_df.append(grouped_df)
    except:
        print('Could not process file.')

concat_df.head()
```

```python
# Just a small check to confirm that bike IDs appear over several monthly files
concat_df.groupby('bikeid').count().sort_values('tripduration').groupby('tripduration').count().head()
```

```python
# Create the final df we're using for analysis
df = concat_df.groupby('bikeid').agg(
    {
        'tripduration': 'sum',
        'first_ride': 'min',
        'last_ride': 'max',
        'tripcount': 'sum'
    }
).reset_index()
df.head()
```

```python
df.dtypes
```

```python
# Add some modified date columns
# This is very hacky because I'm converting to dt each time, need to clean up
df['first_year'] = pd.to_datetime(df['first_ride']).dt.year
df['first_ride_month'] = pd.to_datetime(df['first_ride']).dt.to_period('M')
df['last_ride_month'] = pd.to_datetime(df['last_ride']).dt.to_period('M')
df['first_ride'] = pd.to_datetime(df['first_ride']).dt.date
df['last_ride'] = pd.to_datetime(df['last_ride']).dt.date
df.dtypes
```

```python
# Some more massaging
df['lifespan_days'] = (df['last_ride'] - df['first_ride']).dt.days
df.head()
```

```python
# Tripduration is given in seconds, so we'll convert to hours
df['tripduration_hours'] = df['tripduration'] / (60*60)
df['tripduration_hours'] = df['tripduration_hours'].round(1)
df.head()
```

```python
# Just looking at the age of the fleet
# Didn't add that many bikes in 2018, interesting!

df.groupby('first_year').count()[['bikeid']].rename({'bikeid': 'bike_count'})
```

```python
# My friend was wondering whether the total trip duration of bikes was normally distributed
# It looks a little more bimodal, which might be due to the way of new bikes being added,
# i.e. a lot of bikes in 2017, just a few in 2018, and then again more in 2019/2020,
# so there's fewer "middle-aged" bikes out there
# FYI I'm cutting off the spike at 0 days and the long tail of bikes with a lot of hours

%matplotlib inline
plt.rcParams['figure.figsize'] = [14,10]

df.groupby('tripduration_hours').count()[['bikeid']].rename({'bikeid': 'bike_count'}).plot()
plt.xlim(right=2500)
plt.ylim(top=20)
plt.show()
```

```python
# Look at bike lifespan days and tripduration
# This plot opens in a new window due to the hover feature, so I included a screenshot below
# to be rendered in the HTML

%pylab
import matplotlib.pyplot as plt
from matplotlib import cm
import mplcursors

plt.rcParams['figure.figsize'] = [14,10]

def get_value_at_xy(x, y, column_name):
    row = df.loc[(df['lifespan_days']==x) & (df['tripduration_hours']==y)]
    return row[column_name].values[0]

sizes = pd.DataFrame()
sizes['size'] = df['tripcount'] * 0.1

colors = 1/df['first_year']

fig, ax = plt.subplots()
ax.scatter('lifespan_days', 
           'tripduration_hours', 
           c=colors, 
           cmap='Spectral_r', 
           s=sizes, 
           alpha=0.5, 
           data=df
)
cursor = mplcursors.cursor(ax, hover=True)
cursor.connect("add", lambda sel: sel.annotation.set_text(
    f"""Bike ID: {get_value_at_xy(sel.target[0], sel.target[1], 'bikeid')}
    Total number of trips: {get_value_at_xy(sel.target[0], sel.target[1], 'tripcount')}
    Bike lifespan (days): {sel.target[0]}
    Total trip hours: {sel.target[1]}
    First trip year: {get_value_at_xy(sel.target[0], sel.target[1], 'first_year')}"""
    )
)
plt.xlabel("Bike lifespan in days", size=14)
plt.ylabel("Total trip duration in hours", size=14)
plt.show()
```

## Bike analysis output
This is a bubble plot showing the bike lifespan in days, the total trip duration for a bike in hours, 
the total number of trips, and the first trip year of a bike as follows:
- The x axis shows the bike lifespan in days, ranging from 0 to a little over 1400 days
- The y axis shows the total trip duration per bike in hours, ranging from 0 to a little over 6000 hours
- The size of the bubble represents the total number of trips
- The color represents the year of the first ride of a bike: red=2017, orange=2018, etc.
    
There's a pretty clear trend for bikes to have more hours and a longer lifespan if they're older, of course.
But it's also interesting to see that there are a lot of older bikes that had a fairly short lifespan.
